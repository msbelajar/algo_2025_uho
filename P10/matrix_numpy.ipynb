Aljabar Linear

```python
A = np.array([[1, 3, 1], [1, -2, 0], [1, 1, -1]])
A
```

```python
#determinan
np.linalg.det(A)
```

```python
#invers
np.linalg.inv(A)
```

```python
#transpose
np.linalg.matrix_transpose(A)
```

```python
#nilai eigen dan vektor eigen
np.linalg.eig(A)
```

```python
#pisahkan nilai eigen (E) dan vektor eigen (V)
E, V = np.linalg.eig(A)
```

```python
E
```

```python
V
```

```python
#hanya nilai eigen
np.linalg.eigvals(A)
```

```python
#solusi persamaan linear dalam matriks. Solusi dari Ax = u atau x = inv(A) * u
u = np.array([1, 2, 3])
np.linalg.solve(A,u)
```

```python
np.linalg.inv(A) @ u
```
