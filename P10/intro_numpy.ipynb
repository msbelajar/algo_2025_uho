### Generate Array (membangun array)

```python
#array nol 
x = np.zeros(5)
x
```

```python
#array satu
x = np.ones(7)
x
```

```python
#array random antara 0 dan 1
x = np.random.random(5)
x
```

```python
#array random bilangan bulat antara -5 dan 5 (-5 <= x < 5) -> termasuk -5 tetapi tidak termasuk 5
np.random.randint(-5, 5, 10)
```

```python
#array konstan
x = np.full(5, 8) #ukuran 5 dengan entri 8
x
```

```python
#array dengan suatu permulaan
x = np.arange(1,10,2) #array dimulai dari 1 hingga < 10 dengan penambahan (langkah) 2
x
```

```python
x = np.arange(10,1,-2) #array dimulai dari 10 dengan langkah -2 hingga > 1
x
```

```python
np.arange(10) #otomatis dari 0 hingga < 10
```

```python
np.arange(1,6) #mulai dari 1 hingga <6
```

```python
np.arange(0, 5, 0.5) #mulai dari 0, meningkat 0.5 hingga < 5
```

```python
#array mulai dari 0 hingga 2 (termasuk 2) sebanyak 10 titik
x = np.linspace(0, 2, 10)
x
```

```python
#jika tidak didefinisikan banyak titiknya maka otomatis mengembalikan (default) sebanyak 50 titik
np.linspace(0, 10)
```

```python
#partisi/selang antara a hingga b (a < b) terbagi dalam banyak 5 partisi/selang
a = 0
b = 1
partisi = 5
np.linspace(a, b, partisi+1)
```

### Fungsi Matematika

```python
x = np.random.randint(-10,10, 5)
x
```

```python
#nilai mutlak (absolute)
np.absolute(x)
```

```python
#akar kuadrat
z = np.array([0.4, 1, 9, 15])
np.sqrt(z)
```

```python
#sinus
np.sin(x)
```

```python
#cosinus
np.cos(x)
```

```python
#tangent
np.tan(x)
```

```python
#exponent e^x
np.exp(x)
```

```python
#logaritma natural (ln)
y = np.array([1, 10, 3])
np.log(y)
```

```python
#logaritma basis 10 
y = np.array([1, 10, 3])
np.log10(y)
```
