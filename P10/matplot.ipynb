```python
import numpy as np
import matplotlib.pyplot as plt

#jika belum terinstall matplolib, install dahulu lewat terminal (cmd) dengan mengetikkan pip install matplotlib
```

```python
#grafik sin x dari 0 hingga 2*pi
x = np.linspace(0, 2*np.pi, 100) #jika menggunakan linspace perbanyak titiknya untuk grafik lebih mulus
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#grafik cos x dari 0 hingga 2*pi
x = np.arange(0, 2*np.pi, 0.1) #jika menggunakan arange gunakan delta (kenaikan) yang lebih kecil untuk gambar lebih mulus
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
#fungsi f(x) = x^3 - x^2 + x - 1 pada selang -5 sampai 5
x = np.linspace(-5, 5, 100)
#y = np.pow(x,3) + np.square(x) + x - 1
y = x**3 - x**2 + x - 1
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(np.pi/4, np.pi, 0.01)
u = np.sin(1/(x**2- x))
v = np.cos(1/(x**2- x))
plt.plot(x, u, x, v)
plt.show()
```

```python
x = np.arange(np.pi/4, np.pi, 0.01)
u = np.sin(1/(x**2- x))
v = np.cos(1/(x**2- x))
# plt.figure()
plt.plot(x, u)
plt.plot(x, v)
plt.show()
```

#### Warna (_color_)
Kode warna menggunakan bahasa inggris seperti _blue_ , _red_ , _green_ dsb
Kode warna juga bisa menggunakan kode warna r=red, g=green, b=blue, c=cyan, m=magenta, y=yellow, k=black
warna dapat diatur melalui parameter _color_ pada plot

#### Style garis (_linestyle_)
_Linestyle_ dapat diatur sebagai berikut '-' = solid/garis tak putus, '--' = garis putus-putus, '-.' = garis putus-putus titk dashdot, ':' = titik-titik
linestyle dapat diatur melalui parameter _linestyle_

#### Penanda (_marker_)
'.' = titik bulatan penuh
'o' = lingkaran
'v' = segitiga bawah
'^' = segitiga
'*' = bintang
'x' = silang x

penanda dapat diatur menggunakan parameter _marker_ pada plot

### Cara ringkas
Misalkan ingin membuat garis merah style titik-titik cukup menggunakan 'r:'


```python
x = np.arange(np.pi/4, np.pi, 0.05)
u = np.sin(1/(x**2- x))
v = np.cos(1/(x**2- x))
# plt.figure()
plt.plot(x, u, color='r', linestyle=':')
plt.plot(x, v, 'b--') # 'b--' berarti color=(b)lue , linestyle='--'
plt.show()
```

```python
#lengkapi titel dan label x dan y
x = np.arange(-2*np.pi, 2*np.pi, 0.05)
u = np.sin(x)
v = np.cos(x)
plt.title(r"Grafik sin(x) dan cos(x) pada selang $-2\pi \leq x \leq 2\pi$")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, u,'g--', label="sin(x)")
plt.plot(x, v, 'r', label="cos(x)")
plt.legend() #digunakan untuk menunjukkan label legend
plt.grid() #digunakan untuk menunjukkan grid
plt.show()
```
