```python
from sympy import Symbol
from sympy.plotting import plot
```

```python
x = Symbol('x')
f = x**2 - 3*x - 10
plot(f)
```

```python
plot(f, (x,-5,5))
```

```python
from sympy.plotting import plot3d
from sympy import *
```

```python
y = Symbol('y')
F = sin(x*y)
plot3d(F, (x,-5,5), (y,-5,5))
```
