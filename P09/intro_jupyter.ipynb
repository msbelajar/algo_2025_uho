```python
# menukar nilai
a = 2
b = 5
print(f"a = {a} dan b = {b}")
c = a
a = b
b = c
print(f"a = {a} dan b = {b}")
```

```python
#cara cepat menukar nilai
a = 2
b = 5
print(f"a = {a} dan b = {b}")
a,b = b,a
print(f"a = {a} dan b = {b}")
```
