## Pengurutan (Sorting)

### 1. Selection Sort

```python
A = [89, 45, 68, 90, 29, 34, 17]
n = len(A)
print(f"A = {A}")
for i in range(n-1): #0 sampai n-2
    min = i
    for j in range(i+1,n): #i+1 sampai n-1
        if A[j] < A[min]:
            min = j
    A[i], A[min] = A[min], A[i] #swap 
    print(A)   
```

### 2. Bubble Sort

```python
A = [89, 45, 68, 90, 29, 34, 17]
n = len(A)
print(f"A = {A}")
for i in range(n-1): # 0 sampai n-2
    for j in range(n-1-i): #0 sampai n-2-i
        if A[j+1]<A[j]:
            A[j], A[j+1] = A[j+1], A[j] #swap
    print(A)
```

### 3. Insertion Sort

```python
A = [89, 45, 68, 90, 29, 34, 17]
n = len(A)
print(f"A = {A}")
for i in range(1,n): # 1 sampai n-1
    v = A[i]
    j = i-1
    while j>=0 and A[j]>v:
        A[j+1]=A[j]
        j=j-1
    A[j+1]=v
    print(A)
```
